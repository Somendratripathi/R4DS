### Exploring outliers by size

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, size = class))
# We get a warning here, because mapping an unordered variable (class) to an ordered aesthetic (size) is not a good idea
```

```r
# Left
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, alpha = class))

# Right
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, shape = class))
```

```r
#Here, the color doesn’t convey information about a variable, but only changes the appearance of the plot. 
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy), color = "blue")
```

```r
inr_oil = cbind(inr,oil_prices)
```

```r
inr_oil = inr_oil[,2:ncol(inr_oil)]
inr_oil$Month = as.Date(inr_oil$Month,"%m/%d/%Y")
```

```r
names(inr_oil) = c("INRPrice",
"Open",
"High",
"Low",
"Change",
"Month",
"Oilprice")
```

```r
ggplot(data = subset(inr_oil,Month>as.Date('1988-01-01')))+
    geom_line(mapping = aes(x= Month, y = Oilprice),color ="Blue") +
    geom_line(mapping = aes(x=Month,y = INRPrice),color="Red") +
    ggtitle("INR VS OIL - Time >1988") +
    xlab("Time") + ylab("Currency(INR -Red) / Oil($ - Blue) Value")

ggplot(data = subset(inr_oil,Month>as.Date('2000-01-01')))+
    geom_line(mapping = aes(x= Month, y = Oilprice),color ="Blue") +
    geom_line(mapping = aes(x=Month,y = INRPrice),color="Red") +
    ggtitle("INR VS OIL - Time >2000") +
    xlab("Time") + ylab("Currency(INR -Red) / Oil($ - Blue) Value")


ggplot(data = subset(inr_oil,Month>as.Date('2010-01-01')))+
    geom_line(mapping = aes(x= Month, y = Oilprice),color ="Blue") +
    geom_line(mapping = aes(x=Month,y = INRPrice),color="Red") +
    ggtitle("INR VS OIL - Time >2010") +
    xlab("Time") + ylab("Currency(INR -Red) / Oil($ - Blue) Value")


ggplot(data = subset(inr_oil,Month>as.Date('2015-01-01')))+
    geom_line(mapping = aes(x= Month, y = Oilprice),color ="Blue") +
    geom_line(mapping = aes(x=Month,y = INRPrice),color="Red") +
    ggtitle("INR VS OIL - Time >2015") +
    xlab("Time") + ylab("Currency(INR -Red) / Oil($ - Blue) Value")
    


ggplot(data = subset(inr_oil,Month>as.Date('2018-01-01')))+
    geom_line(mapping = aes(x= Month, y = Oilprice),color ="Blue") +
    geom_line(mapping = aes(x=Month,y = INRPrice),color="Red") +
    ggtitle("INR VS OIL - Time >2018") +
    xlab("Time") + ylab("Currency(INR -Red) / Oil($ - Blue) Value")
```
